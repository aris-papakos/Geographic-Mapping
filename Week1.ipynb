As a first step, we will structure the data into a pandas DataFrame to simplify the data manipulation. We will start by loading the csv file of data into a DataFrame called population_data and we will then filter columns based on our use. *Take a look at the data and its fields*

The complete dataset has 13,572 rows and 15 columns. It can be verified by looking at the shape of your dataframe by <i>population_data.shape</i>. To select specific columns, create a list of column names and view the dataframe as:

```python
# Filtering the data by the columns ['NAME','SEX','AGE','POPEST2016_CIV']
population_data[['NAME','SEX','AGE','POPEST2016_CIV']].head()
```

```python
# Filters out data for all sexes and all age group and store in a new dataframe 'population_data_all'
population_data_all = population_data[population_data['SEX']==0]
population_data_all = population_data_all[population_data_all['AGE']!=999]
```

```python
population_data_all
```

```python
# Sum the population of each state for each year on the dataset 'population_data_all'
population_data_all.groupby(by=['NAME'], as_index=False)[['POPEST2010_CIV','POPEST2011_CIV','POPEST2012_CIV',
                                        'POPEST2013_CIV','POPEST2014_CIV','POPEST2015_CIV','POPEST2016_CIV']].sum()
```

```python
'''Find The Weighted Average - From literature'''

def wavg(group, avg_name, weight_name):
    """ http://stackoverflow.com/questions/10951341/pandas-dataframe-aggregate-function-using-multiple-columns
    In rare instance, we may not have weights, so just return the mean. Customize this if your business case
    should return otherwise. Also reference to http://pbpython.com/weighted-average.html.
    """
    d = group[avg_name]
    w = group[weight_name]
    try:
        return float((d * w).sum()) / w.sum()
    except ZeroDivisionError:
        return d.mean()
    
#Groupby to find the weighted average using the wavg function 
#population_data_all.columns = population_data_all.columns.replace({"NAME":"STATE_NAME"})
avg_age = population_data_all.groupby('NAME').apply(wavg, "AGE", "POPEST2016_CIV")
avg_age = avg_age.reset_index()

avg_age.columns=['STATE_NAME','av_age']
avg_age.head()
```

```python
'Different way of findind the weighted average without  a function.  Our try.'

population_data_all2 = population_data_all
Age = []
for state, statedf in population_data_all2.groupby("NAME"):
    Age.append( (state, sum(statedf["AGE"] * statedf["POPEST2016_CIV"]/statedf["POPEST2016_CIV"].sum()))  )
    
Age = pd.DataFrame(Age)
Age.columns = ["STATE_NAME", "AGE"]
Age.head()
```

```python
#Preparing again the weighted average for the first visualization with matlibplot
avg_age2 = population_data_all.groupby(by=['STATE']).apply(wavg,'AGE','POPEST2016_CIV')
avg_age2.head()
```

<h3> Get a US states shapefile </h3>        
Sources
- ArcGIS shapefile of US 50 states + DC (https://www.arcgis.com/home/item.html?id=f7f805eb65eb4ab787a0a3e1116ca7e5)
<ul>
Recommended – convert to GeoJSON    
(<i> ogr2ogr  -f  GeoJSON  [name1].geojson  [name2].shp </i> - where, 'name1' is the name of the downloaded geojson file and 'name2' is the name you want to specify for the shape file)

```python
#S_DIR is the directory in which your converted name2.shp file is located 
S_DIR = 'US_shape/' 

BLUE = '#5599ff'

with open(os.path.join(S_DIR, 'states.geojson')) as rf:    
    data = json.load(rf)

fig = plt.figure() 
ax = fig.gca()
```

```python
for feature in data['features']:
    geometry = feature['geometry']
    if geometry['type'] == 'Polygon':
        poly = geometry
        ax.add_patch(PolygonPatch(poly, fc=BLUE, ec=BLUE,  alpha=0.5, zorder=2))
    else:
        for polygon in geometry['coordinates'][0]:
            poly = Polygon(polygon)
            ax.add_patch(PolygonPatch(poly, fc=BLUE, ec=BLUE, alpha=0.5, zorder=2))
            

ax.axis('scaled')
plt.axis('off')
plt.show()
```

```python
#Generate colormaps
avg_age2.head()
cmap = cm.get_cmap("hot")

colors = (avg_age2-avg_age2.min())/(avg_age2.max()-avg_age2.min())
colors = colors.apply(lambda x: rgb2hex(cmap(x)))
```

```python
from shapely import affinity
'''Generates a Colormap of weighted age averages of each state using the Matplotlib library.'''

fig = plt.figure(figsize=[15,12]) 
ax = fig.gca()
for feature in data['features']:
    geometry = feature['geometry']
    fips = int(feature['properties']['STATE_FIPS'])
    if geometry['type'] == 'Polygon':
        poly = geometry
        ax.add_patch(PolygonPatch(poly, fc=colors[fips], ec='BLACK',  alpha=0.5, zorder=2))
    else:
        for polygon in geometry['coordinates'][0]:
            color = BLUE
            poly = Polygon(polygon)
            if feature['properties']['STATE_NAME']=='Hawaii':  #scale and move hawaii
                poly = affinity.scale(poly, xfact=25,yfact=25) 
                poly = affinity.translate(poly,xoff=20)
            if feature['properties']['STATE_NAME']=='Alaska': #move alaska
                #poly = affinity.scale(poly, xfact=0.6,yfact=0.6) 
                poly = affinity.translate(poly,xoff=20)
                
            ax.add_patch(PolygonPatch(poly, fc=colors[fips], ec='BLACK', alpha=0.5, zorder=2))

ax.axis('scaled')
plt.axis('off')

#plt.colorbar(ax.get_children()[:len(avg_age)-1],ax=ax)
plt.show()
```

```python
%matplotlib inline
import geopandas as  gpd
import warnings

warnings.filterwarnings('ignore')
    
US = gpd.read_file('US_shape/states.geojson')  #opens the .geojson file as geopandas dataframe
US.sortlevel
US.head()
```

```python
#Plot Geodataframe
US.plot(cmap='OrRd', figsize=(15,15),linewidth=1,edgecolor = "black",legend=True)
plt.title('USA - First Plot',fontsize=16, fontweight='bold')
plt.axis('off')
```

```python
# adds an extra column to the geodataframework for the  weighted average per state.

US=US.merge(avg_age, on="STATE_NAME")
US.head()
```

```python
US['av_age'].head()
```

```python
#create a first projection coloring the map based on the weighted average per state.

US.plot(column='av_age',cmap='OrRd', figsize=(15,15),linewidth=1,edgecolor = "black",scheme='fisher_jenks',legend=True)
plt.title('USA - Colormap',fontsize=16, fontweight='bold')
plt.axis('off')
```

```python
# finds the index of Alaska in to the geo-dataframe
a =US.loc[US['STATE_NAME']=="Alaska"].index[0]
a
```

```python
#convert Alaska to a GeoSeries object in or to srink it

Al = US['geometry'][a]
Al_ser = gpd.GeoSeries(Al) #Convert to GeoSeries

Al_scale=Al_ser.scale(xfact=0.6, yfact=0.6,origin='center') #Scales Alaska

#sc_a=ser.buffer(-0.3)
Al_scale=Al_scale.translate(xoff=14, yoff=-7) #Moves Alaska down to the right
type(Al_scale)
```

```python
Al_scale.plot()
```

```python
from shapely.geometry.multipolygon import MultiPolygon

b=MultiPolygon(Al_scale.all()) #convert again GeoSeries to a multipolygon
US['geometry'][a]=b
```

```python
#plots again the map  with Alaska state shrinked and moved

US.plot(column='av_age',cmap='OrRd', figsize=(15,15),linewidth=1,edgecolor = "black",scheme='fisher_jenks',legend=True)
plt.title('USA - Scaled Alaska',fontsize=16, fontweight='bold')
plt.axis('off')
```

```python
# finds Hawaiiin to the geo-dataframe
h =US.loc[US['STATE_NAME']=="Hawaii"].index[0]
print h
```

```python
#convert Hawaii to a GeoSeries object and stretch it(scale it to a bigger one)

haw = US['geometry'][h]
haw_ser = gpd.GeoSeries(haw) #convert to a Geoseries
haw_ser =haw_ser.scale(xfact=2.0, yfact=2.0,origin='center') #scales Hawaii

#type(sc_a2)
```
